```python
import kagglehub
import shutil
import os

# Define dataset folder
dataset_folder = "datasets/"

# Check if 'datasets/' folder exists, if not create it
if not os.path.exists(dataset_folder):
    os.makedirs(dataset_folder)

# List of datasets to download
datasets = ["reddynitin/aug-train"] # اپنی مطلوبہ datasets کی فہرست یہاں ڈالیں

# Loop through each dataset
for dataset in datasets:
    try:
        print(f"Downloading: {dataset} ...")
        path = kagglehub.dataset_download(dataset)  # Download dataset
        shutil.move(path, dataset_folder)  # Move to datasets folder
        print(f"✅ Successfully downloaded and moved: {dataset}")
    except Exception as e:
        print(f"❌ Error downloading {dataset}: {e}")
```
